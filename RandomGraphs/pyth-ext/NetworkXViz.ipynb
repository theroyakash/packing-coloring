```python
# NetworkX visualization

from rich.console import Console
console = Console()

import networkx as nx
import matplotlib.pyplot as plt

# Generate a random graph with n nodes and probability p of edge creation
n = 10  # Number of nodes
p = 0.46  # Probability of edge creation

G = nx.gnp_random_graph(n, p)

# Add labels to nodes
labels = {node: str(node) for node in G.nodes()}

# Draw the graph
pos = nx.spring_layout(G)  # Layout algorithm (spring_layout is just an example)
nx.draw(G, pos, with_labels=True, labels=labels, node_size=700, node_color='skyblue', font_size=8, font_color='black', font_weight='bold', edge_color='gray')

# Show the plot
plt.show()
```
